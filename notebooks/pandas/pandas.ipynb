**Making changes to the data**

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
```

```python
df
```

```python
df.drop(columns=['Total'])
```

```python
df['Total'] = df.iloc[:,4:10].sum(axis=1)
```

```python
df
```

```python
cols = list(df.columns)
cols
```

```python
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df
```

**Filtering Data**

```python
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Fighting')]
```

```python
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Fighting') & (df['Total'] > 550)]
```

```python
newdf = df.loc[(df['Type 1'] == 'Grass')]
newdf
```

```python
newdf.reset_index(drop=True, inplace=True)
newdf
```

```python
newdf = newdf.reset_index(drop=True)
newdf
```

```python
df.loc[df['Name'].str.contains('Mega')]
```

```python
df.loc[~df['Name'].str.contains('Mega')]
```

```python
import re
df.loc[df['Type 1'].str.contains('Fire|Rock', regex=True)]
```

```python
df.loc[df['Type 1'].str.contains('fire|rock', flags=re.I, regex=True)]
```

```python
df.loc[df['Name'].str.contains('^pi', flags=re.I, regex=True)]
```

```python
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'
df
```

```python
df.loc[df['Type 1'] == 'Flamer', 'Type 1'] = 'Fire'
df
```

```python
df.loc[df['Type 1'] == 'Fire', 'Legendary'] = True
df
```
